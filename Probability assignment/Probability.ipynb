```python
import random
#ques 1
# a. Tossing a coin 10,000 times and calculating the experimental probability of heads and tails.

def toss_coin():
    heads = 0
    tails = 0
    for toss in range (10000):
        output = random.choice(['heads', 'tails'])
        if output == 'heads':
            heads  += 1
        
        else:
            tails += 1

    print(f"experimental probability of heads: {heads/10000}")
    print(f"experimental probability of tails: {tails/10000}")

toss_coin()

# b. Rolling two dice and computing the probability of getting a sum of 7.  

def roll_dice():
    total_rolls = 0
    favourable_outcomes = 0
    for dice1 in range(1,7):
        for dice2 in range(1,7):
            total_rolls += 1
            if(dice1 + dice2 == 7):
                favourable_outcomes += 1

    print(f"probability of dice roll = 7 is : {favourable_outcomes/total_rolls}")

roll_dice()
```

```python

#ques 2. Write a function to estimate the probability of getting at least one "6" in 10 rolls of a fair die.

def probab_of_6():
    outputs = 0
    for roll in range(10):
        outcome = random.randint(1,6)
        if outcome == 6:
            outputs += 1

    print(f"probability of getting atleast 1 six in 10 rolls is : {outputs/10}")

probab_of_6()
```

```python
import numpy as np
import random
#que 4 Generate a sample of size 1000 from a discrete random variable with the following distribution:  
#  - P(X=1) = 0.25  
#  - P(X=2) = 0.35  
#  - P(X=3) = 0.4
#Compute the empirical mean, variance, and standard deviation of the sample.  

#first to generate the sample
def generate_sample():
    size = 1000
    sample = [random.choices([1,2,3], weights = [0.25, 0.35, 0.4]) for _ in range(size)]
    emperical_mean = np.mean(sample)
    emperical_variance = np.var(sample)
    emperical_std_dev = np.std(sample)

    print(f"emperical mean is : {emperical_mean} , variance is : {emperical_variance} and standard deviation is : {emperical_std_dev}")

generate_sample()
```

```python
#5. Simulate 2000 random samples from an exponential distribution with a mean of 5. Visualize the distribution using:  
#  a. A histogram.  
#  b. A probability density function (PDF) overlay.

import numpy as np
import matplotlib.pyplot as plt

samples = np.random.exponential(scale = 5, size = 2000)
plt.figure(figsize=(12, 6))
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g', label='Histogram of Samples')
plt.show()

x = np.linspace(0, max(samples), 1000)
pdf = (1/5) * np.exp(-x/5)
plt.plot(x,pdf, 'r-', lw = 2 , label = 'Theoretical PDF')

plt.title("Exponential Distribution (mean = 5)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()


```

```python
# 6. Simulate the Central Limit Theorem by following these steps  
#  a. Generate 10,000 random numbers from a uniform distribution.  
#  b. Draw 1000 samples of size n = 30.  
#  c. Calculate and visualize the distribution of sample means.  

#a 
sample = np.random.uniform(low = 0, high= 2, size=10000)

#b 
mean_array = np.zeros(1000)
for i in range(1000):
    mean_array[i] = np.random.choice(sample, size=30).mean()

plt.figure(figsize=(12, 6))
plt.hist(mean_array, bins=30, density=True, alpha=0.6, color='b', label='Mean Distribution')
plt.title("Distribution of means")
plt.xlabel('Means')
plt.ylabel('Density')
plt.legend()


#c
plt.figure(figsize=(12, 6))
plt.hist(sample, bins=30, density=True, alpha=0.6, color='g', label='Uniform Sample Distribution')
plt.title("Uniform Distribution")
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

plt.show()
```
