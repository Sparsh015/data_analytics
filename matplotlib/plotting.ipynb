```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
prod_cat = np.array(['Furniture','Technology','Office Supplies'])
sales = np.array([4110451.90, 4744557.50, 3787492])
```

```python
plt.bar(prod_cat,sales)
plt.show()
```

```python
plt.bar(prod_cat,sales, width=0.5, align="center", color= "cyan")
plt.title("Sales across product categories", fontdict={"fontsize" : 15, "fontweight" : 1000, "color" : "blue"})
plt.show()
```

```python
plt.bar(prod_cat,sales, width=0.5, align="center", color= "cyan")
plt.title("Sales across product categories", fontdict={"fontsize" : 15, "fontweight" : 1000, "color" : "blue"})

#adding x y labels
plt.xlabel("\nProduct Category", fontdict={"fontsize" : 15, "fontweight" : 500, "color" : "red"})
plt.ylabel("\nSales", fontdict={"fontsize" : 15, "fontweight" : 500, "color" : "red"})

#modifying ticks
ticks = np.arange(0,6000000,1000000)
labels = ["{}M".format(i//1000000) for i in ticks]
plt.yticks(ticks,labels)

plt.show()
```
