```python
#line chart
import matplotlib.pyplot as plt
import numpy as np
```

```python
months = np.array(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
sales = np.array([241268.76,984637.65,843462.67,235434.67,197639.87,123456.78,456789.12,678901.23,789012.34,890123.45,901234.56,123456.78])
```

```python
plt.plot(months,sales)
plt.title("Monthly Sales Data", fontdict={"fontsize": 15, "fontweight": 1000, "color": "black"})
plt.xlabel("\nMonths", fontdict={"fontsize": 15, "fontweight": 500, "color": "red"})
plt.ylabel("\nSales", fontdict={"fontsize": 15, "fontweight": 500, "color": "red"})
ticks = np.arange(0,1100000, 100000)
labels = ["{}K".format(i//1000) for i in ticks]
plt.yticks(ticks, labels)
plt.xticks(rotation=45)
plt.show()
```

```python
#Box and Whisker Chart

#data
sales_tech = np.array([1012.14, 8398.48,875.51,22320.97,9251.6,4516.65,586.16,174.32,27557.79,563.11,37112.01])
sales_off_supplies = np.array([1770.13,7527.18, 1433.65,432.3,21601.56,10035.73,2378.49,3062.87,38345.71,30345.78])
sales_furniture = np.array([1000.0, 10000.0, 165.0, 243.0, 21000.0, 7300.0, 465.0, 6065.0, 225.0, 29000.0])
```

```python
plt.boxplot([sales_tech, sales_off_supplies, sales_furniture])
#formatting
plt.title("Sales across product categories and countries", fontdict={"fontsize": 15, "fontweight": 1000, "color": "black"})
plt.xlabel("\nProduct categories", fontdict={"fontsize": 15, "fontweight": 500, "color": "red"})
plt.ylabel("\nSales", fontdict={"fontsize": 15, "fontweight": 500, "color": "red"})

#modify ticks
plt.xticks([1, 2, 3], ['Tech', 'Office Supplies', 'Furniture'])
plt.show()
```

```python
#subplots
years = np.array(['2012','2013','2014','2015'])
sales_africa = np.array([123456.78, 234567.89, 345678.90, 456789.01])
sales_Usca = np.array([98765.43, 87654.32, 76543.21, 65432.10])
sales_europe = np.array([54321.09, 43210.98, 32109.87, 21098.76])
sales_asia = np.array([67890.12, 78901.23, 89012.34, 90123.45])
sales_south_america = np.array([12345.67, 23456.78, 34567.89, 45678.90])    
```

```python
fig, ax = plt.subplots()
europe = ax.plot(years, sales_europe, label='Europe')
africa = ax.plot(years, sales_africa, label='Africa')
asia = ax.plot(years, sales_asia, label='Asia')
usa = ax.plot(years, sales_Usca, label='USA')
south_america = ax.plot(years, sales_south_america, label='South America')
africa[0].set_dashes([10,2,5,2])

plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(ncols=3, nrows=2)

europe = ax[0, 1].plot(years, sales_europe, label='Europe')
africa = ax[0][2].plot(years, sales_africa, label='Africa')
asia = ax[0][0].plot(years, sales_asia, label='Asia')
usa = ax[1][0].plot(years, sales_Usca, label='USA')
south_america = ax[1][1].plot(years, sales_south_america, label='South America')
africa[0].set_dashes([10,2,5,2])

#to add subtitles and supertitles of plots andsubplots
plt.suptitle("Sales Data by Region", fontdict={"fontsize": 15, "fontweight": 1000, "color": "black"})
ax[0][0].set_title("Asia", fontdict={"fontsize": 12, "fontweight": 500, "color": "blue"})
ax[0][1].set_title("Europe", fontdict={"fontsize": 12, "fontweight": 500, "color": "blue"})
ax[0][2].set_title("Africa", fontdict={"fontsize": 12, "fontweight": 500, "color": "blue"})
ax[1][0].set_title("USA", fontdict={"fontsize": 12, "fontweight": 500, "color": "blue"})
ax[1][1].set_title("South America", fontdict={"fontsize": 12, "fontweight": 500, "color": "blue"})

#to adjust padding of subplots
fig.tight_layout()

plt.legend()
plt.show()
```
