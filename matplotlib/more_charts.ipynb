```python
#line chart
import matplotlib.pyplot as plt
import numpy as np
```

```python
months = np.array(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
sales = np.array([241268.76,984637.65,843462.67,235434.67,197639.87,123456.78,456789.12,678901.23,789012.34,890123.45,901234.56,123456.78])
```

```python
plt.plot(months,sales)
plt.title("Monthly Sales Data", fontdict={"fontsize": 15, "fontweight": 1000, "color": "black"})
plt.xlabel("\nMonths", fontdict={"fontsize": 15, "fontweight": 500, "color": "red"})
plt.ylabel("\nSales", fontdict={"fontsize": 15, "fontweight": 500, "color": "red"})
ticks = np.arange(0,1100000, 100000)
labels = ["{}K".format(i//1000) for i in ticks]
plt.yticks(ticks, labels)
plt.xticks(rotation=45)
plt.show()
```
