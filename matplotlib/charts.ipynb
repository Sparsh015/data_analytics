```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

sales = np.array([
    1012.4, 465.4, 875.47, 567.34, 102.6, 567.9, 878.98, 6745.09, 3456.98, 76.9, 5.6,      # Tech
    1200.5, 800.2, 950.1, 400.0, 300.3, 600.6, 700.7, 2000.0, 1500.5, 250.2, 100.1,         # Furniture
    900.0, 450.5, 700.7, 350.3, 200.2, 500.5, 600.6, 1800.0, 1200.2, 150.1, 80.0            # Office Supplies
])
profit = np.array([
    -345.67, 1435.6, 894.4, -345.6, -123.5, 345.65, 453.5, -45.9, 5007.8, -98.78, 675.8,    # Tech
    200.0, 300.5, 150.2, -100.1, 50.0, 400.4, 250.2, 800.0, 600.6, 100.1, 20.0,             # Furniture
    100.0, 200.2, 120.1, -50.0, 30.0, 250.5, 180.8, 700.0, 400.4, 60.0, 10.0                # Office Supplies
])
prod_cat = np.array(
    ['Tech']*11 + ['Furniture']*11 + ['Office Supplies']*11
)
countries = np.array(
    ['India','Zimbawe','africa','America','Russia','China','italy','France','India','Zimbawe','africa']*3
)
```

```python
plt.scatter(profit, sales, alpha=0.7, s=50)  # Scatter plot with profit vs sales
```

```python
plt.scatter(profit,sales, color= "blue")
plt.title("Sales across product categories", fontdict={"fontsize" : 15, "fontweight" : 1000, "color" : "black"})

#adding x y labels
plt.xlabel("\nProduct Category", fontdict={"fontsize" : 15, "fontweight" : 500, "color" : "red"})
plt.ylabel("\nSales", fontdict={"fontsize" : 15, "fontweight" : 500, "color" : "red"})
```

```python
plt.scatter(profit[prod_cat == "Tech"], sales[prod_cat == "Tech"], c = "green", s = 50, alpha=0.7, label="Tech")
plt.scatter(profit[prod_cat == "Furniture"], sales[prod_cat == "Furniture"], c = "blue", s = 50, alpha=0.7, label="Furniture")
plt.scatter(profit[prod_cat == "Office Supplies"], sales[prod_cat == "Office Supplies"], c = "yellow", s = 50, alpha=0.7, label="Office Supplies")
plt.legend()
plt.title("Sales vs product categories across product categories and countries", fontdict={"fontsize" : 15, "fontweight" : 1000, "color" : "black"})

#adding x y labels
plt.xlabel("\nProfit", fontdict={"fontsize" : 15, "fontweight" : 500, "color" : "red"})
plt.ylabel("\nSales", fontdict={"fontsize" : 15, "fontweight" : 500, "color" : "red"})
```
